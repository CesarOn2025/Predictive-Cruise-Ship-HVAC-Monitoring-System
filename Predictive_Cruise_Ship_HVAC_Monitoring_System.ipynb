```python
import random

# =========================================
# HVAC SENSOR SIMULATION
# =========================================

temperature = random.randint(65, 95)

vibration = round(random.uniform(0.1, 5.0), 2)

pressure = random.randint(18, 40)

# =========================================
# DISPLAY SENSOR VALUES
# =========================================

print("HVAC TEMPERATURE:", temperature, "F")

print("HVAC VIBRATION:", vibration)

print("HVAC PRESSURE:", pressure, "PSI")
```

```python
# =========================================
# PREDICTIVE MAINTENANCE LOGIC
# =========================================

if temperature > 85 and vibration > 3.5:
    status = "HIGH RISK - Possible compressor failure"

elif temperature > 80:
    status = "WARNING - High temperature detected"

elif vibration > 3.0:
    status = "WARNING - Excessive vibration detected"

elif pressure < 20:
    status = "WARNING - Low pressure detected"

else:
    status = "SYSTEM NORMAL"
    print("\nSYSTEM STATUS:")

print(status)
```

```python
# =========================================
# NLP MAINTENANCE REPORT ANALYSIS
# =========================================

maintenance_report = (
    "Unit 4 shows abnormal vibration and possible overheat condition."
)

keywords = [
    "abnormal",
    "vibration",
    "failure",
    "overheat",
    "shutdown"
]

detected = []

for word in keywords:
    if word in maintenance_report.lower():
        detected.append(word)

# =========================================
# NLP RESULTS
# =========================================

print("\nMAINTENANCE REPORT:")

print(maintenance_report)

print("\nNLP ANALYSIS:")

if detected:
    print("Potential issues detected:", ", ".join(detected))
else:
    print("No critical issues detected")
```

```python
# =========================================
# DATA STORAGE AND REPORTING
# =========================================

import pandas as pd

from datetime import datetime

results = {
    "Timestamp": datetime.now(),
    "Temperature": temperature,
    "Vibration": vibration,
    "Pressure": pressure,
    "System Status": status,
    "NLP Analysis": ", ".join(detected)
}

# CREATE DATAFRAME

df = pd.DataFrame([results])

# DISPLAY TABLE

print("\n===== HVAC PREDICTIVE MAINTENANCE REPORT =====\n")

print(df)

# SAVE CSV REPORT

df.to_csv("hvac_report.csv", index=False)

print("\nCSV report saved successfully.")
```

```python
# =========================================
# HVAC DASHBOARD VISUALIZATION
# =========================================

import matplotlib.pyplot as plt

# SENSOR VALUES

sensor_names = ["Temperature", "Vibration", "Pressure"]

sensor_values = [temperature, vibration, pressure]

# CREATE BAR CHART

plt.figure(figsize=(8,5))

plt.bar(sensor_names, sensor_values)

# TITLES

plt.title("HVAC System Monitoring Dashboard")

plt.ylabel("Sensor Values")

# SHOW GRAPH

plt.show()
```

```python
# =========================================
# HISTORICAL HVAC DATA LOGGING
# =========================================

import os

history_file = "hvac_history.csv"

# CHECK IF FILE EXISTS

if os.path.exists(history_file):

    existing_data = pd.read_csv(history_file)

    updated_data = pd.concat([existing_data, df], ignore_index=True)

else:
    updated_data = df

# SAVE UPDATED HISTORY

updated_data.to_csv(history_file, index=False)

print("\nHistorical HVAC data updated successfully.")

print("\nTotal Records Stored:")

print(len(updated_data))
```

```python
# =========================================
# HVAC TREND ANALYSIS
# =========================================

import matplotlib.pyplot as plt

# LOAD HISTORICAL DATA

history_df = pd.read_csv("hvac_history.csv")

# CREATE TEMPERATURE TREND GRAPH

plt.figure(figsize=(10,5))

plt.plot(history_df["Temperature"], marker='o')

# TITLES

plt.title("HVAC Temperature Trend")

plt.xlabel("System Runs")

plt.ylabel("Temperature (F)")

# SHOW GRAPH

plt.show()
```

```python
# =========================================
# HVAC RISK SCORING SYSTEM
# =========================================

risk_score = 0

# TEMPERATURE RISK

if temperature > 85:
    risk_score += 40

elif temperature > 80:
    risk_score += 20

# VIBRATION RISK

if vibration > 3.5:
    risk_score += 40

elif vibration > 3.0:
    risk_score += 20

# PRESSURE RISK

if pressure < 20:
    risk_score += 20

# =========================================
# RISK CLASSIFICATION
# =========================================

if risk_score >= 80:
    risk_level = "CRITICAL"

elif risk_score >= 50:
    risk_level = "HIGH"

elif risk_score >= 20:
    risk_level = "MODERATE"

else:
    risk_level = "LOW"

# =========================================
# DISPLAY RESULTS
# =========================================

print("\n===== HVAC RISK ANALYSIS =====\n")

print("Risk Score:", risk_score)

print("Risk Level:", risk_level)
```

```python
# =========================================
# MULTI-UNIT HVAC MONITORING
# =========================================

import pandas as pd
import random

hvac_units = []

# CREATE MULTIPLE HVAC UNITS

for unit in range(1, 6):

    temperature = random.randint(65, 95)

    vibration = round(random.uniform(0.1, 5.0), 2)

    pressure = random.randint(18, 40)

    # RISK SCORE

    risk_score = 0

    if temperature > 85:
        risk_score += 40

    elif temperature > 80:
        risk_score += 20

    if vibration > 3.5:
        risk_score += 40

    elif vibration > 3.0:
        risk_score += 20

    if pressure < 20:
        risk_score += 20

    # RISK LEVEL

    if risk_score >= 80:
        risk_level = "CRITICAL"

    elif risk_score >= 50:
        risk_level = "HIGH"

    elif risk_score >= 20:
        risk_level = "MODERATE"

    else:
        risk_level = "LOW"

    # STORE UNIT DATA

    hvac_units.append({
        "Unit": f"HVAC-{unit}",
        "Temperature": temperature,
        "Vibration": vibration,
        "Pressure": pressure,
        "Risk Score": risk_score,
        "Risk Level": risk_level
    })

# CREATE TABLE

multi_df = pd.DataFrame(hvac_units)

# DISPLAY RESULTS

print("\n===== MULTI-UNIT HVAC MONITORING =====\n")

print(multi_df)
```

```python
# =========================================
# EXECUTIVE HVAC RISK DASHBOARD
# =========================================

import matplotlib.pyplot as plt

# BAR CHART

plt.figure(figsize=(10,5))

plt.bar(multi_df["Unit"], multi_df["Risk Score"])

# TITLES

plt.title("HVAC Risk Score Dashboard")

plt.xlabel("HVAC Units")

plt.ylabel("Risk Score")

# SHOW GRAPH

plt.show()
```

```python
# =========================================
# REAL-TIME HVAC ALERT SYSTEM
# =========================================

print("\n===== HVAC ALERT CENTER =====\n")

for index, row in multi_df.iterrows():

    if row["Risk Level"] == "HIGH":

        print(
            f"WARNING ALERT: {row['Unit']} requires maintenance attention."
        )

    elif row["Risk Level"] == "CRITICAL":

        print(
            f"CRITICAL ALERT: {row['Unit']} may experience failure soon."
        )

    else:

        print(
            f"{row['Unit']} operating normally."
        )
```

```python
# =========================================
# HVAC MAINTENANCE PRIORITY RANKING
# =========================================

priority_df = multi_df.sort_values(
    by="Risk Score",
    ascending=False
)

print("\n===== MAINTENANCE PRIORITY RANKING =====\n")

print(
    priority_df[
        [
            "Unit",
            "Risk Score",
            "Risk Level"
        ]
    ]
)
```

```python
# =========================================
# FAILURE PROBABILITY ESTIMATION
# =========================================

# CALCULATE FAILURE PROBABILITY

multi_df["Failure Probability (%)"] = (
    multi_df["Risk Score"] * 1.25
)

# LIMIT MAXIMUM TO 100%

multi_df["Failure Probability (%)"] = (
    multi_df["Failure Probability (%)"]
    .clip(upper=100)
)

# DISPLAY RESULTS

print("\n===== FAILURE PROBABILITY ANALYSIS =====\n")

print(
    multi_df[
        [
            "Unit",
            "Risk Score",
            "Failure Probability (%)"
        ]
    ]
)
```

```python
# =========================================
# MAINTENANCE RECOMMENDATION ENGINE
# =========================================

recommendations = []

# ANALYZE EACH HVAC UNIT

for index, row in multi_df.iterrows():

    if row["Risk Level"] == "CRITICAL":

        recommendation = (
            "Immediate inspection and shutdown recommended."
        )

    elif row["Risk Level"] == "HIGH":

        recommendation = (
            "Schedule maintenance within 24 hours."
        )

    elif row["Risk Level"] == "MODERATE":

        recommendation = (
            "Monitor closely and inspect during next maintenance cycle."
        )

    else:

        recommendation = (
            "Continue normal operation."
        )

    recommendations.append(recommendation)

# ADD RECOMMENDATIONS COLUMN

multi_df["Maintenance Recommendation"] = recommendations

# DISPLAY RESULTS

print("\n===== MAINTENANCE RECOMMENDATIONS =====\n")

print(
    multi_df[
        [
            "Unit",
            "Risk Level",
            "Maintenance Recommendation"
        ]
    ]
)
```

```python
# =========================================
# EXECUTIVE KPI DASHBOARD
# =========================================

# TOTAL HVAC UNITS

total_units = len(multi_df)

# CRITICAL UNITS

critical_units = len(
    multi_df[multi_df["Risk Level"] == "CRITICAL"]
)

# HIGH RISK UNITS

high_units = len(
    multi_df[multi_df["Risk Level"] == "HIGH"]
)

# AVERAGE RISK SCORE

average_risk = round(
    multi_df["Risk Score"].mean(),
    2
)

# DISPLAY KPI DASHBOARD

print("\n===== EXECUTIVE HVAC KPI DASHBOARD =====\n")

print(f"Total HVAC Units: {total_units}")

print(f"Critical Units: {critical_units}")

print(f"High Risk Units: {high_units}")

print(f"Average Risk Score: {average_risk}")
```

```python
# =========================================
# EXPORT EXECUTIVE HVAC REPORT
# =========================================

report_columns = [
    "Unit",
    "Temperature",
    "Vibration",
    "Pressure",
    "Risk Score",
    "Risk Level",
    "Failure Probability (%)",
    "Maintenance Recommendation"
]

# CREATE EXECUTIVE REPORT

executive_report = multi_df[report_columns]

# EXPORT CSV

executive_report.to_csv(
    "executive_hvac_report.csv",
    index=False
)

# DISPLAY RESULTS

print("\n===== EXECUTIVE REPORT GENERATED =====\n")

print(executive_report)

print("\nFile saved: executive_hvac_report.csv")
```

```python
# =========================================
# HVAC WORK ORDER GENERATION
# =========================================

work_orders = []

# GENERATE WORK ORDERS

for index, row in multi_df.iterrows():

    if row["Risk Level"] in ["HIGH", "CRITICAL"]:

        work_orders.append({
            "Unit": row["Unit"],
            "Priority": row["Risk Level"],
            "Action": row["Maintenance Recommendation"]
        })

# CREATE WORK ORDER TABLE

work_order_df = pd.DataFrame(work_orders)

# DISPLAY RESULTS

print("\n===== GENERATED WORK ORDERS =====\n")

print(work_order_df)
```
